## FOR ALL BIRDS

```python
datazone_df = pd.DataFrame(columns=[
    "common_name", "scientific_name", "family", 
    "image_src", "page_url"
])
for bird in birds:
    html_content = str(bird)
    soup = BeautifulSoup(html_content, 'html.parser')
    try:
        img_source = soup.find('img')['src']
    except:
        img_source = "No image found"
    
    bird_data = {
        "common_name": soup.find('h5').get_text(),
        "scientific_name": soup.find('p', class_='italic').get_text(),
        "family": soup.find('p', class_='text-default-text-secondary').get_text(),
        "image_src": img_source,
        "page_url": "https://datazone.birdlife.org/" + soup.find('a')['href']
    }
    
    # print(json.dumps(bird_data, indent=4))
    datazone_df = pd.concat([datazone_df, pd.DataFrame([bird_data])], ignore_index=True)

datazone_df
```

```python
datazone_df[['common_name', 'scientific_name']].to_csv("datazone_birdlist.csv", index=False)
```
